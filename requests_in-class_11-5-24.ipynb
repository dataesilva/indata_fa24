```python
airport_endpoint = 'https://aviationweather.gov/api/data/airport'
```

```python
query = '?ids={}&format={}'.format('@OH', 'json')
```

```python
airport_response = requests.get(airport_endpoint + query)
airport_response.status_code
```

```python
airport_dat = json.loads(airport_response.text)
airport_df = pd.DataFrame(airport_dat)
airport_df.head()
```
