```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
header = {'User-agent': 'Mozilla/5.0'}
jobs = requests.get('https://www.usajobs.gov/Search/Results?j=1560&l=Kent%2C%20Ohio&l=Kent%2C%20Ohio&k=Data%20Science%20Series&p=1', headers = header)
```

```python
jobs.status_code
```

```python
BeautifulSoup(jobs.text)
```

```python
with open('jobs_usa.html', 'w') as text:
    text.write(jobs.text)
```
