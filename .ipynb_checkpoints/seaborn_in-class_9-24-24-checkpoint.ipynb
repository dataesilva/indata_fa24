### Using hex codes for colors instead of names

```python
sns.relplot(data = t_clean,
            x = 'age',
            y = 'fare',
           alpha = 0.5,
           hue = 'sex',
           col = 'survived',
           palette = {'male': '#3A0CA3', 'female':'#67ff33'})

plt.show()
```

```python
t_clean['Sex'] = t_clean['sex'].map({'male': 'Male', 'female':'Female'})
t_clean['Sex'].iloc[0:5]
```

```python
sns.relplot(data = t_clean,
            x = 'age',
            y = 'fare',
           alpha = 0.5,
           hue = 'Sex', # Change here
           col = 'survived',
           palette = {'Male': 'blue', 'Female':'magenta'}) # Change here also

plt.show()
```

```python
sns.relplot(data = t_clean,
            x = 'age',
            y = 'fare',
           alpha = 0.5,
           hue = 'Sex', # Change here
           col = 'survived',
           palette = {'Male': 'blue', 'Female':'magenta'},
           size = 'pclass') # Change here also

plt.show()
```

#### This step reverses the 1 - 3 values so that the upper class (1) is larger on the final figure

```python
t_clean['Passenger Class'] = 4 - t_clean['pclass']
t_clean['Passenger Class'].iloc[0:5]
```

### Final challenge figure completed

```python
t_clean['Survival Status'] = t_clean['survived'].map({0: 'Died', 1:'Survived'})
t_clean['Survival Status'].iloc[0:5]
```

```python
sns.relplot(data = t_clean,
            x = 'fare',
            y = 'age',
           alpha = 0.5,
           hue = 'Sex',
           col = 'Survival Status',
           palette = {'Male': '#3A0CA3', 'Female':'#67ff33'},
           size = 'Passenger Class')

plt.show()
```

### Group 2

```python
mh = pd.read_csv('Student Mental health.csv')
mh.head()
```

```python
mh.groupby('Choose your gender')[['Do you have Depression?',
                                   'Do you have Anxiety?',
                                  'Do you have Panic attack?']].count()
```

```python
columns_to_plot = mh.columns[-4:-1]
category_order = ['Yes', 'No']
```

```python
columns_to_plot
```

```python
for column in columns_to_plot:
    counts = mh[column].value_counts()
    
    counts = counts.reindex(category_order, fill_value = 0)
    print(counts)
```

```python
mh['Do you have Depression?'].value_counts()
```

```python
sns.countplot(data = mh,
             x = 'Do you have Depression?',
             stat = 'percent',
             hue = 'Did you seek any specialist for a treatment?')

# Not included in the demo
plt.show()
```

```python
sns.countplot(data = mh,
             x = 'Do you have Anxiety?',
             stat = 'percent',
             hue = 'Did you seek any specialist for a treatment?')

# Not included in the demo
plt.show()
```

#### Conceptual question, what is the value of choosing which of these to map to the x-axis and to map to the hue? These are the decisions a data display designer needs to be able to answer

```python
sns.countplot(data = mh,
             hue = 'Do you have Anxiety?',
             stat = 'percent',
             x = 'Did you seek any specialist for a treatment?')

plt.show()
```

```python
sns.countplot(data = mh,
             x = 'Do you have Panic attack?',
             stat = 'percent',
             hue = 'Did you seek any specialist for a treatment?')

plt.show()
```

```python
sns.countplot(data = mh,
             x = 'Do you have Depression?',
             stat = 'percent',
             hue = 'Choose your gender',
             palette = 'PuBuGn')

plt.show()
```

#### This for loop is not required in the challenge, but applies the concept used earlier in the challenge to the drawing of the figures.

```python
for column in columns_to_plot:
    sns.countplot(data = mh,
             x = column,
             stat = 'percent',
             hue = 'Choose your gender',
             palette = 'PuBuGn')
    plt.show()
```
